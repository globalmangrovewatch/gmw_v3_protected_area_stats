```python
import pandas
import rsgislib.tools.utils
```

```python
input_file = "Mangrove in WDPA July 2022.xlsx"
```

```python
cntry_stats_df = pandas.read_excel(input_file)
```

```python
cntry_stats_df["Country"] = ""
cntry_stats_df["UNRegion"] = ""
cntry_stats_df["ReportRegion"] = ""
```

```python
cntry_stats_df = cntry_stats_df[cntry_stats_df["ISO_TER"] != "GLOBAL"]
```

```python
cntry_stats_df
```

```python
cntry_names_lut_file = "gadm_lut.json"
cntry_unregion_lut_file = "country_unregions_lut.json"
cntry_rep_region_lut_file = "country_report_regions_lut.json"
```

```python
cntry_names_lut = rsgislib.tools.utils.read_json_to_dict(cntry_names_lut_file)
cntry_unregion_lut = rsgislib.tools.utils.read_json_to_dict(cntry_unregion_lut_file)
cntry_rep_region_lut = rsgislib.tools.utils.read_json_to_dict(cntry_rep_region_lut_file)
```

```python
cntry_unregion_rvs_lut = dict()

for rgn in cntry_unregion_lut:
    for cntry in cntry_unregion_lut[rgn]:
        cntry_unregion_rvs_lut[cntry] = rgn
```

```python
cntry_report_region_rvs_lut = dict()

for rgn in cntry_rep_region_lut:
    for cntry in cntry_rep_region_lut[rgn]:
        cntry_report_region_rvs_lut[cntry] = rgn
```

```python
for index, row in cntry_stats_df.iterrows():
    cntry_stats_df.at[index,'Country'] = cntry_names_lut["gid"][row["ISO_TER"]]
    cntry_stats_df.at[index,'UNRegion'] = cntry_unregion_rvs_lut[cntry_names_lut["gid"][row["ISO_TER"]]]
    cntry_stats_df.at[index,'ReportRegion'] = cntry_report_region_rvs_lut[cntry_names_lut["gid"][row["ISO_TER"]]] 
```

```python
cntry_stats_df
```

```python
report_region_stats_df = cntry_stats_df.groupby("ReportRegion").sum()
```

```python
report_region_stats_df
```

```python
report_region_stats_df["PERCENT PROTECTED"] = (report_region_stats_df["MANGROVE AREA PROTECTED(HA)"] / report_region_stats_df["TOTAL MANGROVE AREA(HA)"])*100


report_region_stats_df["MANGROVE AREA PROTECTED(km2)"] = (report_region_stats_df["MANGROVE AREA PROTECTED(HA)"] / 100)
report_region_stats_df["TOTAL MANGROVE AREA(km2)"] = (report_region_stats_df["TOTAL MANGROVE AREA(HA)"] / 100)
```

```python
report_region_stats_df = report_region_stats_df.reset_index()
report_region_stats_df.to_feather("protected_area_proportions_regions.feather")
report_region_stats_df.to_csv("protected_area_proportions_regions.csv")
excel_sheet = 'protect_regions'
xls_writer = pandas.ExcelWriter("protected_area_proportions_regions.xlsx", engine='xlsxwriter')
report_region_stats_df.to_excel(xls_writer, sheet_name=excel_sheet)
xls_writer.save()
```
